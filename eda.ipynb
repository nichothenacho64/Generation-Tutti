```python
# from pathlib import Path
import pathlib
from typing import Optional

def read_conversation(name: str):
    DIRECTORY = pathlib.Path("kiparla-data/kipasti-data")  # Change this if needed

    if "csv" in name:
        file_type = "*.csv" 
    elif "tsv" in name:
        file_type = "*.tsv" 
    else:
        return None

    for csv_file in DIRECTORY.glob(file_type):
        file_name = str(csv_file)
        if name and name in file_name:
            conversation = pd.read_csv(
                csv_file,
                sep="\t", # the data is tab delimited, not comma-separated in this case
                engine="python",
                quotechar='"',
                keep_default_na=False,
                on_bad_lines="warn" # just in case
            )
            if "vert" in file_name:
                conversation = conversation.rename(columns={
                                                "span" : "with_jefferson",
                                                "form" : "without_jefferson",
                                                })
            # else:
            #     conversation = conversation.drop(["tu_id"], axis=1) # the text unit id is the index, so it is not needed...

            return conversation
    else:
        print("No conversation found")
```
